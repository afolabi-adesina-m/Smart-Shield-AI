## **1. EXECUTIVE SUMMARY**
The **Ontario Smart-Shield** is an advanced AI system designed to improve road safety by predicting real-time highway hazards (e.g., black ice, whiteouts, accidents). While standard GPS systems rely on static speed limits, this system generates a **Dynamic Speed Recommendation** by fusing unstructured data (Images and Text) with structured environmental data. The project demonstrates a complete ML lifecycle, from multimodal feature extraction to optimized logistic decision-making.

## **2. THE PROBLEM STATEMENT**
Navigation apps like Google Maps effectively track traffic volume but fail to account for **instantaneous environmental hazards** common in Ontario.
*   **The Safety Gap:** Drivers are often warned of a "slowdown" but not the *cause* (e.g., black ice), leading to high-speed entries into dangerous zones.
*   **The Technical Challenge:** Traditional linear regression is too simple to capture the non-linear relationship between wind, visibility, and road surface texture.

## **3. TECHNICAL METHODOLOGY (The 3-Brain Architecture)**

### **Pillar 1: The NLP Brain (The Social Listener)**
*   **Action:** Scrapes and parses real-time text alerts from **Ontario 511** and **OPP** feeds.
*   **Technique:** Uses an NLP Pipeline (Cleaning, Stop-word removal, and Stemming) and **TF-IDF Encoding** to turn emergency keywords into a numeric **Textual Risk Index ($T$)**.

### **Pillar 2: The Vision Brain (The Visual Sensor)**
*   **Action:** Connects to live **Hwy 401/QEW camera APIs**. 
*   **Technique:** Uses **OpenCV** to normalize images and a **Neural Network (CNN)** to identify road textures (e.g., "White" = Snow, "Reflective" = Ice) to produce a **Visual Hazard Index ($V$)**.

### **Pillar 3: The Logistic Optimizer (The Decision Engine)**
*   **Action:** Fuses all data points into a final prediction.
*   **Model:** **Logistic Regression** optimized via **GridSearchCV**.
*   **Feature Selection:** Uses **Lasso (L1)** to identify if weather sensors or visual cameras are the more significant predictor in specific conditions.

## **4. SAFETY SCORE LOGIC**
The system calculates an **Integrated Safety Score ($S$)** using a weighted decision matrix:

$$S = (w_v \cdot V) + (w_t \cdot T) + (w_e \cdot E)$$

*   **$V$ (Visual Index):** Real-time texture classification (0–1).
*   **$T$ (Textual Index):** Keyword severity score from 511 alerts (0–1).
*   **$E$ (Environmental Index):** Numeric data (Visibility in km, Temperature, Wind).
*   **$w$ (Weights):** Dynamically assigned based on model confidence.

### **The Decision Rule:**
*   **Score < 30:** **GREEN** (Safe) - Maintain 100/110 km/h.
*   **Score 30–70:** **YELLOW** (Caution) - Recommend 80 km/h.
*   **Score > 70:** **RED** (Danger) - Recommend 60 km/h or stop.
